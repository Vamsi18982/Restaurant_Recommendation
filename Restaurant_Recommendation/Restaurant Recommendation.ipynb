```python
import pandas as pd

# Read the CSV file into a DataFrame
data = pd.read_csv("test_100k.csv")
```

```python
data.head()
```

```python
# Drop columns
columns_to_drop = ['customer_id', 'gender','vendor_category_en','delivery_available','language','restaurent_tag_name']  # L,ist of columns to drop
data = data.drop(columns=columns_to_drop)

```

```python
data.head()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
sns.pairplot(data)
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state = 678)
    kmeans.fit(data)
    cluster.append((i, kmeans.inertia_))
```

```python
cluster
```

```python
plt.plot(cluster)
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 45)

kmeans.fit(data)
```

```python
kmeans.predict([data.iloc[5, :]])
```

```python
data.iloc[5, :]
```

```python
labels = kmeans.labels_
```

```python
labels
```

```python
plt.scatter(data["serving_distance"],data["restaurant_id"], c = labels)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
lin = linkage(data[["serving_distance","restaurant_id"]][:10])
```

```python
lin
```

```python
dendrogram(lin)
plt.show()
```
